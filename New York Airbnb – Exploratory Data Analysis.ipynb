# Comparing price values with Neighbourhood_group

```python
sns.barplot(data=df1,x='neighbourhood_group',y='price')
plt.title('Price comparision with Neighbourhood group')
plt.show()
```

****Adding further drill of Room type for detail analysis****

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df1,x='neighbourhood_group',y='price',hue='room_type')
plt.title('Price comparision with Neighbourhood group with room types')
plt.show()
```
